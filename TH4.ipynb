```python
#4A2
n = 0
for i in range(1000):
    n = int(input())
    if n > 0:
        break
    else:
        print("Nhap lai")
if n == 1:
    print("1 không phải là số nguyên tố.")
else:
    prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            prime = False
    if prime:
        print(n, "là số nguyên tố.")
    else:
        print(n, "không phải là số nguyên tố.")
```

```python
#4A3
n = int(input("Nhập n (0 < n < 100): "))
if 0 < n < 100:
    result = 1
    for i in range(1, n + 1):
        result *= i
    print(f"{n}! = {result}")
else:
    print("Giá trị n không hợp lệ.")
```

```python
#4A4
n = int(input("Nhập số nguyên n: "))

if n < 0:
    n = -n

if n == 0:
    count = 1
else:
    count = 0
    while n > 0:
        n //= 10
        count += 1

print("Số chữ số là:", count)
```
