```python
#4A4
n = int(input("Nhập số nguyên n: "))

if n < 0:
    n = -n

if n == 0:
    count = 1
else:
    count = 0
    while n > 0:
        n //= 10
        count += 1

print("Số chữ số là:", count)
```

```python
#4A6
a, b = map(int, input().split())

sum_primes = 0

num = a
while num <= b:
    is_prime_num = True
    
    if num < 2:
        is_prime_num = False
    else:
        i = 2
        while i*i <= num:
            if num % i == 0:
                is_prime_num = False
                break
            i += 1
            
    if is_prime_num:
        sum_primes += num
        
    num += 1

print(sum_primes)
```

```python
#4A7
n = int(input())

largest_prime = 0
temp_n = n
i = 2

while i * i <= temp_n:
    if temp_n % i == 0:
        largest_prime = i
        while temp_n % i == 0:
            temp_n //= i
    i += 1

if temp_n > 1:
    largest_prime = temp_n

print(largest_prime)
```

```python
#4A8
n = int(input())
steps = 0

while True:
    
    s_n = str(n)
    is_palindrome = True
    i = 0
    j = len(s_n) - 1
    while i < j:
        if s_n[i] != s_n[j]:
            is_palindrome = False
            break
        i += 1
        j -= 1
        
    if is_palindrome:
        break
        
    reversed_n = 0
    temp_n = n
    while temp_n > 0:
        reversed_n = reversed_n * 10 + temp_n % 10
        temp_n //= 10
        
    n = n + reversed_n
    steps += 1

print(steps, n)
```
